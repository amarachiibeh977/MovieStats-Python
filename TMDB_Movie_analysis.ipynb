```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
```

```python
high_budgeted_years = moviedata.groupby('release_year')['budget_adj'].sum().sort_values(ascending=False)
plt.figure(figsize =(10,6))
plt.plot(high_budgeted_years,color = 'skyblue')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.xlabel('release_year')
plt.ylabel('budjet_adj')
plt.title('Highest budgeted years')
plt.show
```

```python
most_budget_adjusted_year = moviedata.groupby('release_year')['budget_diff'].sum().sort_values(ascending=False)
plt.figure(figsize =(10,6))
plt.plot(most_budget_adjusted_year)
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.xlabel('release_year')
plt.ylabel('budjet_diff')
plt.title('Most budget adjusted year')
plt.show
```

```python
high_budgeted_movies = moviedata.groupby(['release_year', 'original_title'])['budget_adj'].sum().sort_values(ascending=False).head(10)

# Reset index to access 'release_year' and 'original_title' as columns
high_budgeted_movies = high_budgeted_movies.reset_index()

# Combine year and title for better labeling
high_budgeted_movies['movie_label'] = high_budgeted_movies['release_year'].astype(str) + " - " + high_budgeted_movies['original_title']

# Create a horizontal bar chart
plt.figure(figsize=(12, 6))
plt.barh(high_budgeted_movies['movie_label'], high_budgeted_movies['budget_adj'], color='skyblue')

# Labels & title
plt.xlabel('Adjusted Budget')
plt.ylabel('Movies (Year - Title)')
plt.title('Top 10 Highest Budgeted Movies')
plt.gca().invert_yaxis()  # Invert to show highest budget at the top

plt.show()
```

```python
top_revenue_movies = moviedata.groupby(['release_year', 'original_title'])['revenue_adj'].sum().sort_values(ascending=False).head(10)
top_revenue_movies = top_revenue_movies.reset_index()
top_revenue_movies['movie_label'] = top_revenue_movies['release_year'].astype(str)+ '-' +top_revenue_movies['original_title']
plt.figure(figsize=(12,6))
plt.barh(top_revenue_movies['movie_label'], top_revenue_movies['revenue_adj'], color='skyblue')
plt.xlabel('Revenue')
plt.ylabel('Movies (Year - Title)')
plt.title('Top 10 revenue generated movies')
plt.gca().invert_yaxis()
plt.show()

```

```python
top_profit_movies = moviedata.groupby(['release_year', 'original_title'])['profit'].sum().sort_values(ascending=False).head(10)
top_profit_movies = top_profit_movies.reset_index()
top_profit_movies['movie_label'] = top_profit_movies['release_year'].astype(str)+ '-' +top_profit_movies['original_title']
plt.figure(figsize=(12,6))
plt.barh(top_profit_movies['movie_label'], top_profit_movies['profit'], color='skyblue')
plt.xlabel('Profit')
plt.ylabel('Movies (Year - Title)')
plt.title('Top 10 profit generated movies')
plt.gca().invert_yaxis()
plt.show()
```

```python
top_profit_genre = moviedata.groupby('genres')['profit'].sum().head(4).sort_values(ascending=False)
plt.figure(figsize=(10,6))
plt.barh(top_profit_genre.index, top_profit_genre)
plt.xlabel('Profit')
plt.xlabel('Genre')
plt.title('Top 4 profit generated genres')

plt.gca().invert_yaxis()
plt.show
```

```python
top_genres = moviedata['genres'].value_counts().head(10)

# Seaborn bar plot (works like a histogram for categorical data)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.index, y=top_genres.values, palette='coolwarm')

# Labels & title
plt.xlabel('Genres')
plt.ylabel('Count')
plt.title('Top 10 Most Common Movie Genres')
plt.xticks(rotation=45)  # Rotate for readability

plt.show()
```

```python
corr_matrix = moviedata[['budget', 'revenue', 'vote_average', 'vote_count', 'popularity_level_code']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Movie Data")

plt.show()
```

```python
plt.figure(figsize=(12, 10))

# Scatter plot: Budget vs Revenue
plt.subplot(2, 2, 1)
sns.scatterplot(x=moviedata['budget'], y=moviedata['revenue'], color='royalblue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs Revenue')

# Scatter plot: Revenue vs Vote Count
plt.subplot(2, 2, 2)
sns.scatterplot(x=moviedata['revenue'], y=moviedata['vote_count'], color='teal')
plt.xlabel('Revenue')
plt.ylabel('Vote Count')
plt.title('Revenue vs Vote Count')

# Scatter plot: Vote Count vs Vote Average
plt.subplot(2, 2, 3)
sns.scatterplot(x=moviedata['vote_count'], y=moviedata['vote_average'], color='darkorange')
plt.xlabel('Vote Count')
plt.ylabel('Vote Average')
plt.title('Vote Count vs Vote Average')

# Scatter plot: Budget vs Vote Count
plt.subplot(2, 2, 4)
sns.scatterplot(x=moviedata['budget'], y=moviedata['vote_count'], color='purple')
plt.xlabel('Budget')
plt.ylabel('Vote Count')
plt.title('Budget vs Vote Count')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
moviedata['release_month'] = pd.to_datetime(moviedata['release_date']).dt.month
```

```python
monthly_revenue = moviedata.groupby('release_month')['revenue'].sum().sort_values(ascending=False)
monthly_profit = moviedata.groupby('release_month')['profit'].sum().sort_values(ascending=False)
```

```python
# Ensure 'release_date' is in datetime format
moviedata['release_date'] = pd.to_datetime(moviedata['release_date'])

# Extract Month
moviedata['release_month'] = moviedata['release_date'].dt.month

# Group by Month and Calculate Total Revenue & Profit
monthly_revenue = moviedata.groupby('release_month')['revenue'].sum()
monthly_profit = moviedata.groupby('release_month')['profit'].sum()

plt.figure(figsize=(10, 5))

sns.lineplot(x=monthly_revenue.index, y=monthly_revenue.values, marker='o', label='Revenue', color='blue')
sns.lineplot(x=monthly_profit.index, y=monthly_profit.values, marker='s', label='Profit', color='green')

plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xlabel('Month')
plt.ylabel('Total Revenue & Profit')
plt.title('Monthly Revenue & Profit Trends')
plt.legend()
plt.grid(True)

plt.show()
```
